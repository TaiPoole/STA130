Simple linear regression is a way to try to approximate an output value y, using x, and a combination of other values. The formula is basically just the formula of a line: y = beta0 * x + beta1 + error. You have a "starting point" (beta1), an amount of change of the output based on the input (beta0), and then an amount of potential error that is based on the normal distribution (sigma). To form a sample, use the line equation to make a true line, and then add potential variance so that it has a spread around the regression line based on the sigma value.

````python
import numpy as np
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt

beta_0 = 3  
beta_1 = 5 
sigma = 2   

x_values = np.linspace(0, 10, 50)

y_true_values = beta_0 * x_values + beta_1

error = stats.norm.rvs(loc=0, scale=sigma, size=x_values.shape)

y_observed_values = y_true_values + error

plt.figure(figsize=(10, 6))
sns.lineplot(x=x_values, y=y_true_values, color='blue', label="True regression line (y_true)", linewidth=2)
sns.scatterplot(x=x_values, y=y_observed_values, color='red', alpha=0.6, label="Observed data points (y_observed)")
````

> _**R-squared** is the "the proportion of variation in (outcome) $Y$ explained by the model ($\hat y_i$)" and is defined as_
>

>
> _The visuzation provided in the previous problem can be used to consider $(Y_i-\bar Y)^2$ as the squared distance of the $Y_i$ to their sample average $\bar Y$ as opposed to the squared **residuals** $(Y_i-\hat y)^2$ which is the squared distance of the $Y_i$ to their fitted (predicted) values $Y_i$._

### 13. ### (A) Do a permuation test $\;H_0: \mu_{\text{short}}=\mu_{\text{long}} \; \text{ no difference in duration between short and long groups}$ by "shuffling" the labels
### (B) Create a 95% bootstrap confidence interval  by repeatedly bootstrapping within each group and applying *np.quantile(bootstrapped_mean_differences, [0.025, 0.975])* to the collection of differences between the sample means.

````python
mean_short = old_faithful[old_faithful['kind'] == 'short']['duration'].mean()
mean_long = old_faithful[old_faithful['kind'] == 'long']['duration'].mean()
observed_diff = mean_long - mean_short

permuted_diffs = []

for _ in range(10000):
    shuffled_labels = np.random.permutation(old_faithful['kind'])
    mean_short_perm = old_faithful[shuffled_labels == 'short']['duration'].mean()
    mean_long_perm = old_faithful[shuffled_labels == 'long']['duration'].mean()
    permuted_diff = mean_long_perm - mean_short_perm
    permuted_diffs.append(permuted_diff)

permuted_diffs = np.array(permuted_diffs)
p_value = np.mean(np.abs(permuted_diffs) >= np.abs(observed_diff))
print(f"p-value: {round(p_value, 4)}")
````

Here we can see that by shuffling the labels and comparing if the means are different, we get strong evidence to reject the null hypothesis that "there is no difference". This means that the labels are placed meaningfully and that there is definitely a relationship in waiting and duration time.

````python
short_durations = old_faithful[old_faithful['kind'] == 'short']['duration']
long_durations = old_faithful[old_faithful['kind'] == 'long']['duration']

bootstrapped_mean_differences = []

for _ in range(10000):
    short_sample = np.random.choice(short_durations, size=len(short_durations), replace=True)
    long_sample = np.random.choice(long_durations, size=len(long_durations), replace=True)
    
    mean_diff = long_sample.mean() - short_sample.mean()
    bootstrapped_mean_differences.append(mean_diff)

confidence_interval = np.quantile(bootstrapped_mean_differences, [0.025, 0.975])

print(f"95% Bootstrap Confidence Interval for the mean difference: {confidence_interval}")
````

Here we can see that our confidence interval ranges from 2.1 to 2.3, suggesting that you can say with high confidence (95% confidence...) that there is a difference between the categories.

Histograms of Residuals: We visualized the residuals (differences between observed and predicted values) from the models, exploring how the residual distributions compare to a normal distribution. The goal was to evaluate the assumption that the error terms (residuals) are normally distributed, which is a key assumption for linear regression.

Boxplots: We also used boxplots (via Plotly) to visualize the distribution of duration across different values of kind ("short" vs "long"), providing insight into the difference in the groups.

6. Comparison of Statistical Methods
Permutation vs. Bootstrap vs. Indicator Variable Based Model:
Permutation Test and Bootstrap CI are non-parametric methods, relying on randomness to generate distributions of statistics and make inferences about group differences.
The Indicator Variable Model is a parametric method, estimating the effect of the kind variable on duration through a linear regression model.
The permutation test provides a p-value, the bootstrap test provides a confidence interval, and the indicator variable model provides a regression coefficient and a p-value for the group differences.
The permutation test and bootstrap confidence interval offer more flexibility as they make fewer assumptions about the data, whereas the indicator variable model assumes a linear relationship between variables and relies on specific statistical assumptions.

> The first three assumptions associated with the **Simple Linear regression** model are that
> 
> - the $\epsilon_i$ **errors** (sometimes referred to as the **noise**) are **normally distributed**
> - the $\epsilon_i$ **errors** are **homoscedastic** (so their distributional variance $\sigma^2$ does not change as a function of $x_i$)
> - the linear form is [at least reasonably approximately] "true" (in the sense that the above two remain [at least reasonably approximately] "true") so that then behavior of the $Y_i$ **outcomes** are represented/determined on average by the **linear equation**)<br>
> 
>    and there are additional assumptions; but, a deeper reflection on these is "beyond the scope" of STA130; nonetheless, they are that<br><br>
> - the $x_i$ **predictor variable** is **measured without error**
> - and the $\epsilon_i$ **errors** are **statistically independent** (so their values do not depend on each other)
> - and the $\epsilon_i$ **errors** are **unbiased** relative to the **expected value** of **outcome** $E[Y_i|x_i]=\beta_0 + \beta_1x_i$ (which is equivalently stated by saying that the mean of the **error distribution** is $0$, or again equivalently, that the **expected value** of the **errors** $E[\epsilon_i] = 0$)
