````python
import numpy as np

# Function to perform bootstrapping and calculate confidence intervals
def bootstrap_confidence_interval(data, num_bootstrap_samples=1000, confidence_level=0.95):
    # Number of data points
    n = len(data)
    
    # Array to store the bootstrapped means
    bootstrapped_means = np.zeros(num_bootstrap_samples)
    
    # Generate bootstrap samples and calculate their means
    for i in range(num_bootstrap_samples):
        # Resample with replacement
        bootstrap_sample = np.random.choice(data, size=n, replace=True)
        # Calculate the mean of the bootstrap sample
        bootstrapped_means[i] = np.mean(bootstrap_sample) # IF WE WANT MEDIAN CHANGE THIS TO MEDIAN OR WHATEVER POPULATION PARAMETER YOU WANT
    
    # Sort the bootstrapped means
    sorted_means = np.sort(bootstrapped_means)
    
    # Calculate the confidence interval percentiles
    lower_bound = np.percentile(sorted_means, (1 - confidence_level) / 2 * 100)
    upper_bound = np.percentile(sorted_means, (1 + confidence_level) / 2 * 100)
    
    return lower_bound, upper_bound

# Example usage
data = [10, 12, 13, 15, 18, 21, 23, 25, 27, 29]  # Sample data
lower, upper = bootstrap_confidence_interval(data)

print(f"95% confidence interval: [{lower:.2f}, {upper:.2f}]")
````

## Quantitative analysis

To explore the effect of the vaccine/drug (or whether it even has one), we will use bootstrapping to examine the confidence intervals of our data.

````python
np.random.seed(314)    

observed_diff = np.mean(df['ChangeInHealthScore'])

def bootstrap_difference(treatment, n_bootstrap=100000):
    boot_diffs = []
    for _ in range(n_bootstrap):
        boot = np.random.choice(treatment, size=len(treatment), replace=True) #size = population, and replace is true, the two key factors of bootstrapping
        mean = np.mean(boot)
        boot_diffs.append(mean) #add the mean to a list so you can make a distribution of them
    
    return np.array(boot_diffs)

bootstrap_diffs = bootstrap_difference(df['ChangeInHealthScore'])

#grab upper and lower bounds
ci_lower = np.percentile(bootstrap_diffs, 2.5)
ci_upper = np.percentile(bootstrap_diffs, 97.5)

#output results
print(f"Observed difference in means: {observed_diff:.2f}")
print(f"95% Confidence Interval: {ci_lower:.2f} - {ci_upper:.2f}")
````

````python
fig=px.histogram(x=bootstrap_diffs)
fig.add_vline(x=observed_diff, line_width=3, line_dash="dash", annotation_text="Mean Difference")
fig.add_vline(x=ci_lower, line_width=2, line_dash="dash", line_color="red")
fig.add_vline(x=ci_upper, line_width=2, line_dash="dash", line_color="red")
fig.show(renderer='png')
#the red lines show the 95% confidence interval
````
